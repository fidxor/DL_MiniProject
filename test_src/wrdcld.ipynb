```python
#pip install transformers
from transformers import BertTokenizer, TFBertModel
```

```python
# BERT 모델 및 토크나이저 로드
model_name = 'bert-base-uncased'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = TFBertModel.from_pretrained(model_name)

# 임베딩할 텍스트
text = "Hello, how are you?"

# 텍스트를 토큰화하고 BERT 입력 형식으로 변환
tokens = tokenizer.encode_plus(text, add_special_tokens=True, padding='max_length', max_length=128, truncation=True, return_tensors='tf')

# BERT 입력 형식으로 변환된 토큰을 모델에 입력하여 임베딩 얻기
outputs = model(tokens.input_ids, attention_mask=tokens.attention_mask)

# 임베딩 결과
embeddings = outputs.last_hidden_state
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

columns = ['E/I', 'S/N', 'T/F', 'J/P']
# 서브 플롯 사이즈
fig, axes = plt.subplots(2, 4, figsize=(15, 5))


for i, col in enumerate(tqdm(columns)):
    # 각각의 columns 에 대해서 값이 0 , 1 일때 총 9000개의 데이터
    filtered_df_0 = df[df[col] == 0][:9000]
    filtered_df_1 = df[df[col] == 1][:9000]
    
    # 단어 필터링
    text_0 = ' '.join(filtered_df_0['posts'].astype(str).tolist())
    text_1 = ' '.join(filtered_df_1['posts'].astype(str).tolist())
    
    # wordcloud (값이 0일때) 에 위치 설정
    wordcloud_0 = WordCloud().generate(text_0)
    axes[i // 4, i % 4].imshow(wordcloud_0, interpolation='bilinear')
    axes[i // 4, i % 4].set_title(col + ' = 0')
    axes[i // 4, i % 4].axis('off')
    # wordcloud (값이 1일때) 에 위치 설정
    wordcloud_1 = WordCloud().generate(text_1)
    axes[(i+4) // 4, (i+4) % 4].imshow(wordcloud_1, interpolation='bilinear')
    axes[(i+4) // 4, (i+4) % 4].set_title(col + ' = 1')
    axes[(i+4) // 4, (i+4) % 4].axis('off')
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

def preprocess_text(text):

    tokens = word_tokenize(text)
    text = text.lower()


    stop_words = stopwords.words('english')
    stop_words.extend(["MBTI", "INFJ", "INTJ", "INFP", "INTP", "ENFJ", "ENTJ", "ENFP", "ENTP", "ISFJ", "ISTJ", "ISFP", "ISTP", "ESFJ", "ESTJ", "ESFP", "ESTP", "personality", "type"])

    tokens = [word for word in tokens if word not in stop_words]
    
    return tokens

filtered_df_0['processed_posts'] = filtered_df_0['posts'].apply(preprocess_text)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = ' '.join(filtered_df_0['processed_posts'].astype(str).tolist())


wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)


plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
